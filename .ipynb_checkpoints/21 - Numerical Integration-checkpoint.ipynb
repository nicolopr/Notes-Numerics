# Numerical integration

## Riemann integrals

The first approximation we are going to use is the usual $\int_{I} f(x)dx=\sum_{i=1}^N hf(x_i)$, where $\{x_i\}_{i=1}^N$ are points splitting the segment $I$ into $N-1$ segments of length $h$.

The total accuracy of this method is $O(h)$, in fact we have that for every interval:

$\int_{x_i}^{x_{i+1}} f(x)dx=\int_{x_i}^{x_{i+1}}(f(x_i)+f'(x_i)(x-x_i)+O(h^2))dx=f(x_i)h+f'(x_i)\bigg(\dfrac{x_{i+1}^2}{2}-\dfrac{x_{i}^2}{2}+x_i^2-x_ix_{i+1}\bigg)+hO(h^2)= hf(x_i)+\dfrac{h^2}{2}f'(x_i)+hO(h^2)=hf(x_i)+O(h^2)$

So every term of the sum gives a $O(h^2)$ error, thus overall we have $nO(h^2)$. Since $h=\dfrac{(b-a)}{n}$, we get $O(h)$ total error.

The __midpoint rule__ can improve on this. Taking a Taylor approximation of $f$ around $y_i\equiv\dfrac{x_{i+1}+x_i}{2}$ we obtain:
$\int_{x_i}^{x_{i+1}} f(x)dx=\int_{x_i}^{x_{i+1}}\bigg(f(y_i)+f'(y_i)(x-y_i)+f''(y_i)\dfrac{(x-y_i)^2}{2}+O(h^3)\bigg)dx$.

The point here is that the odd terms in $(x-y_i)$ are all zero due to symmetry. Hence, redoing all the previous steps we get that:

$\int_{I} f(x) = \sum_{i=1}^N h f(y_i) + O(h^2)$, i.e. an extra order of accuracy.

```python
#example of the two different approaches
import numpy as np

a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_riemannL = h * sum(f[:n-1])
err_riemannL = 2 - I_riemannL

I_riemannR = h * sum(f[1::])
err_riemannR = 2 - I_riemannR

I_mid = h * sum(np.sin((x[:n-1] \
        + x[1:])/2))
err_mid = 2 - I_mid

print(I_riemannL)
print(err_riemannL)

print(I_riemannR)
print(err_riemannR)

print(I_mid)
print(err_mid)
```

## Trapezoid rule

This rule fits a trapezoid in each interval and sums their areas. Each trapezoid will have vertices at:
<center>$(x_i,0),\, (x_{i+1},0), \,(x_i,y_i), \,(x_{i+1}, \,y_{i+1})$</center>

The approximation is then: $\int_{a}^b f(x)dx=\sum_{i=0}^{n-1} h \dfrac{f(x_{i})+f(x_{i-1})}{2}$.

Apparently this double counts many terms and a more accurate approximation is given by:

<center> $\int_{a}^b f(x)dx=\sum_{i=0}^{n-1} \dfrac{h}{2}\big(f(x_0)+ 2\sum_{i=1}^{n-1}f(x_{i})+f(x_n)\big)$</center>

The __accuracy__ is determined as follows. First, Taylor expand  $f(x)$ around the midpoint $y_i=\dfrac{x_{i+1}+x_i}{2}$:

$f(x)\sim f(y_i)+f'(y_i)(x-y_i)+f''(y_i)\dfrac{(x-y_i)^2}{2!}+\dots$.

Then evaluate at $x=x_i$ and $x=x_{i+1}$, and average the two results recalling that $x_i-y_i=-\frac{h}{2}$ and $x_{i+1}-y_i=\frac{h}{2}$:
$f(x_i)=f(y_i)-\frac{h}{2}f'(y_i)+h^2f''(y_i)\dfrac{(x-y_i)^2}{8}+\dots$

$f(x_{i+1})=f(y_i)+\frac{h}{2}f'(y_i)+h^2f''(y_i)\dfrac{(x-y_i)^2}{8}+\dots$.

Therefore:

$\dfrac{f(x_{i+1})+f(x_i)}{2}=f(y_i)+O(h^2)$.

Now go back to the integral over an interval and Taylor expand inside of it:

$\int_{x_i}^{x_{i+1}} f(x)dx=\int_{x_i}^{x_{i+1}}f(y_i)dx+\int_{x_i}^{x_{i+1}}f'(y_i)(x-y_i)dx+\int_{x_i}^{x_{i+1}}f''(y_i)\dfrac{(x-y_i)^2}{2}dx$

Since (x-y_i) is antisymmetric in the interval, odd powers drop out and integrating one gets:

$\int_{x_i}^{x_{i+1}} f(x)dx=hf(y_i)+O(h^3)$

Substituting the expression for $f(y_i)$ derived above we obtain:

<center>$\int_{x_i}^{x_{i+1}} f(x)dx = h\left( \dfrac{f(x_i)+f(x_{i+1})}{2}\right)+O(h^3)$</center>

Therefore summing all intervals we get $O(h^2)$ accuracy.

```python
import numpy as np

a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_trap = (h/2)*(f[0] + \
          2 * sum(f[1:n-1]) + f[n-1])
err_trap = 2 - I_trap

print(I_trap)
print(err_trap)
```

## Simpson's rule

Considering two consecutive intervals $(x_{i-1},x_i)$ and $(x_i, x_{i+1})$, Simpson's rule approximates $f(x)$ by fitting a quadratic polynomial through __three__ points: $(x_{i-1},f(x_{i-1}))$, $(x_{i},f(x_{i}))$,  $(x_{i+1},f(x_{i+1}))$, and integrates the polynomial exactly on the interval $(x_{i-1},x_{i+1})$.

To build the polynomials, we use Lagrange interpolation to obtain:

$P_i(x)=f(x_{i-1})\dfrac{(x-x_{i})(x-x_{i+1})}{(x_{i-1}-x_i)(x_{i-1}-x_{i+1})}+f(x_{i})\dfrac{(x-x_{i-1})(x-x_{i+1})}{(x_{i}-x_{i-1})(x_{i}-x_{i+1})}+f(x_{i+1})\dfrac{(x-x_{i-1})(x-x_{i})}{(x_{i+1}-x_{i-1})(x_{i+1}-x_{i})}$

And substituting for $h$ we get:

$P_i(x)=\dfrac{f(x_{i-1})}{2h^2}(x-x_i)(x-x_{i+1})-\dfrac{f(x_{i})}{h^2}(x-x_{i-1})(x-x_{i+1})+\dfrac{f(x_{i+1})}{2h^2}(x-x_i)(x-x_{i-1})$

The integration is immediate and gives:

$\int_a^b f(x)dx=\dfrac{h}{3}\left[f(x_0)+4\sum_{i \text{ odd}}f(x_i)+2\sum_{i \text{ even}}f(x_i)+f(x_n)\right]$

The __accuracy__ can be computed as usual and is $O(h^4)$, so much better!

```python
a=0
b = np.pi
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)
I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 2 - I_simp
print(I_simp)
print(err_simp)
```

# Computing integrals in Python

`scipy.integrate` has several functions to evaluate integrals. `trapz` takes as input an array of functions values `f` evaluated on a grid.

```python
import numpy as np
from scipy.integrate import trapz
a=0
b = np.pi
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)
I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1]) #manual computation
print(I_trapz)
print(I_trap)
#only differ by numerical precision
```

Sometimes we need to know the primitive $F(X)=\int_{x_0}^x f(x)dx$. This can be done via the `cumtrapz` function of the same package.

```python
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

x = np.arange(0, np.pi, 0.01)
F_exact = -np.cos(x)
F_approx = cumtrapz(np.sin(x), x)
plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title("$F(x) = \int_0^{x} sin(y) dy$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["Exact with Offset", "Approx"])
plt.show()
```

`scipy.integrate` has other integration methods, for example `quad` that takes as arguments `f,a,b`, where `a` and `b` are endpoints of interval. It returns the integral and the estimated error.


```python
from scipy.integrate import quad

I_quad, est_err_quad = quad(np.sin, 0, np.pi)
print(I_quad)
err_quad = 2 - I_quad
print(est_err_quad, err_quad)
```

# Problems

Write a function my_fourier_coef(f,n) with output [An,Bn], where f is an function object that is 2π -periodic. The function my_fourier_coef should compute the nth Fourier coefficients, An and Bn, in the Fourier series for f defined by the two formulas given earlier. Use the quad function to perform the integration.

```python
def my_fourier_coef(f,n):
    Af=lambda x:1/np.pi*f(x)*np.cos(n*x)
    Bf=lambda x:1/np.pi*f(x)*np.sin(n*x)
    An, _=quad(Af,-np.pi,np.pi)
    Bn, _=quad(Bf,-np.pi,np.pi)
    return [An,Bn]
```

```python
def plot_results(f, N):
    x = np.linspace(-np.pi, np.pi, 10000) 
    [A0, B0] = my_fourier_coef(f, 0)
    y = A0*np.ones(len(x))/2
    for n in range(1, N):
        [An, Bn] = my_fourier_coef(f, n)
        y += An*np.cos(n*x)+Bn*np.sin(n*x)
    plt.figure(figsize = (10,6))
    plt.plot(x, f(x), label = "analytic")
    plt.plot(x, y, label = "approximate")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.legend()
    plt.title(f"{N}th Order Fourier Approximation")
    plt.show()
```

```python
f = lambda x: np.sin(np.exp(x))
N=2
plot_results(f, N)
```

```python
f = lambda x: np.mod(x, np.pi/2) 
N=5
plot_results(f, N)
```

```python
N = 20
plot_results(f, N)
```
