# 2D Plots via MatplotLib

```python
plt.plot([1,2,3,4],[10,20,30,40])
plt.show()
```

```python
x = np.linspace(-5,5, 100)
print(len(x)) #discretisation of interval -5, 5 in 100 points
plt.plot(x, x**2, 'g^') #third argument sets the line color and shape. First letter = color name, second letter = line shape.
plt.show()
```

You can add as many plots as you want in the same figure.

```python
x = np.linspace(-7,7, 20)
plt.plot(x, x**2, 'rh') #red triangles
plt.plot(x, x**3, 'co') #cyan circles
plt.plot(x, math.e **x, 'y-') # 'yellow dashes'
plt.show()
```

We can __change size__ to the figure via the method `figsize`. 

We can add __axis descriptions__ and __title__ to any graph using the methods `xlabel`, `ylabel` and `title`.

We can add a __legend__ via the method `legend`. To work properly, you need to have passed `label` parameters to the `plot` methods.

We can show more/less of the plot via the methods `xlim` and `ylim`.

```python
plt.figure(figsize = (10,6))

x = np.linspace(-5,5,20)
plt.plot(x, x**2, 'ko', label='quadrato')
plt.plot(x, x**3, 'r*', label = 'cubo')
plt.title(f'Plot of Various Polynomials from {x[0]} to {x[-1]}')
plt.xlabel('X axis', fontsize = 18)
plt.ylabel('Y axis', fontsize = 18)
plt.legend(loc = 2)
plt.xlim(-10) #no data before -5!
plt.ylim(-10)
plt.show()
```

Predefined styles are contained in the attribute `plt.style.available`.

They can be used via `plt.style.use(style_to_use)`.

## Other types of plots

Brief list of other plots supported by matplotlib:
- `plt.plot` is a __line__ plot.
- `plt.scatter` is a __scatter__ plot.
- `plt.bar` is a __bar__ plot.
- `plt.loglog` is plot with __log__ scale on `x` and `y`.
- `plt.semilogx` is plot with __log__ scale on `x` only.
- `plt.semilogy` is plot with __log__ scale on `y` only.
- `plt.hist` makes an __histogram__.
- `plt.pie` makes a __pie__ chart.

## Utility functions

You can __save__ a figure using `plt.savefig(path, format=format)` in the chosen format (pdf, jpg and others are supported).

`plt.errorbar` returns plots with error bars.

# 3D Plots via mplot3d

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

`mplot3d` lets us add a z-axis to plots. This is done via the `axes` method.

As usual, you can rename it via the `set_zlabel` method. To plot, use `plot3D` method.

```python
fig=plt.figure(figsize=(3,3))
ax=plt.axes(projection= '3d')
t=np.arange(0,10,0.01)
x=np.sin(t)
y=t
ax.plot3D(x,y,t)
#ax.grid()
#ax.set_title('spirale')
#ax.set_xlabel('pippo')
#ax.set_ylabel('baudo')
#ax.set_zlabel('enrico_papi') #not sure why it is cut out
plt.show()
```

We can also create scatter plots in 3d by the usual `scatter` method, after we have added our third axis with `axes`.

```python
fig=plt.figure(figsize=(3,3))
sc=plt.axes(projection = '3d')
x=np.random.random(100) #100 random number between 0 and 1
y=np.random.random(100)
z=np.random.random(100)
sc.scatter(x,y,z,c = 'r', s = 10)
plt.show()
```

Finally, we can make __surface plots__, i.e. graphs of a function $f(x,y)$, via `plot_surface`. First however samples of pair of points $X,Y$ must be created via the `numpy` function `np.meshgrid`.

```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.arange(-5, 5.1, 0.2)
y = np.arange(-5, 5.1, 0.2)

X, Y = np.meshgrid(x, y)
Z = np.sin(X)*np.cos(Y)

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
```

### <font color = red> 4) Plot surface of equation $z=\frac{xy(x^2-y^2)}{x^2+y^2}$ for $x \in (-4,4)$ and $y \in (-3,3)$  </font>

```python
fig=plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
x=np.arange(-4,4,4/100)
y=np.arange(-3,3,3/100)
X,Y = np.meshgrid(x,y)
Z= X*Y*(X**2-Y**2)/(X**2+Y**2)
ax.plot_surface(X,Y,Z, color='g')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_wireframe(X,Y,Z, color='r')
plt.show()
```

### <font color = red> 5) Plot sierpinsky points  </font>

```python
def my_sierpinsky(no_pts):
    fig=plt.figure(figsize=(10,10))
    P1=(0,0)
    P2=(0.5, 2**(-0.5))
    P3=(1,0)
    p1=P1
    count=0
    x=[]
    y=[]
    def find_midpoint(A,B):
        return ((A[0]+B[0])/2, (A[1]+B[1])/2)
    while count<no_pts:
        x.append(p1[0])
        y.append(p1[1])
        randno=np.random.randint(1,4)
        match randno:
            case 1:
                p1=find_midpoint(p1,P1)
            case 2:
                p1=find_midpoint(p1,P2)
            case 3:
                p1=find_midpoint(p1,P3)
        count+=1
    plt.scatter(x,y, s=0.2,c='blue')
    plt.show()
# much faster to do like this rather than plotting each point separately. I guess calling plt.scatter many times is memory expensive.
```

```python
my_sierpinsky(100000)
```

### <font color = red> 5) Plot a fractal  </font>

```python
def fractal_plot(no_pts):
    fig=plt.figure(figsize=(10,10))
    x=[0]
    y=[0]
    count=0
    while count<no_pts:
        rand_no=np.random.randint(1,101)
        if rand_no==1:
            x.append(0)
            y.append(0.16*y[-1])
        elif rand_no<=8:
            x.append(0.2*x[-1] - 0.26*y[-1])
            y.append(0.23*x[-1]+0.22*y[-1]+1.6)
        elif rand_no<=15:
            x.append(-0.15*x[-1] + 0.28*y[-1])
            y.append(0.26*x[-1]+0.24*y[-1]+0.44)
        else:
            x.append(0.85*x[-1] + 0.04*y[-1])
            y.append(-0.04*x[-1]+0.85*y[-1]+1.6)
        count+=1
    plt.axis('equal') 
    plt.axis('off')
    plt.scatter(x,y,s=0.5,c='b')
    plt.show()
```

```python
fractal_plot(10000)
```

### <font color = red> 5) Parametric 3d plotter  </font>

```python
def parametric_plotter(f1,f2,t):
    fig=plt.figure(figsize=(10,10))
    ax=plt.axes(projection='3d')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('t')
    x=[f1(el) for el in t]
    y=[f2(el) for el in t]
    ax.plot3D(x,y,t)
    plt.title('Parametric plot')
    plt.show()
```

```python
f = lambda t: np.sin(t)
g = lambda t: t**2
parametric_plotter(f, g, np.linspace(0, 6*np.pi, 100))
```
